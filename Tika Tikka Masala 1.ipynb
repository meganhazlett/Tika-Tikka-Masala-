# Import and Clean the Data 

```python
tika.initVM()
from tika import parser
tikka1 = parser.from_file('./tiki1.pdf')
tikka1c= tikka1["content"]
print(tikka1c)
```

```python
#Lower case (avoids making useless words "important" because capitalized)
tikka1cL=tikka1c.lower()
#Remove punctuation
tokenizer = RegexpTokenizer(r'\w+')
tikka1cT=tokenizer.tokenize(tikka1cL)
#Keep only useful words
useful_tikka1 = [word for word in tikka1cT if word not in stopwords.words('english')]
#Print 
print(useful_tikka1)
```

```python
#Keep only the nouns 
tags = nltk.pos_tag(useful_tikka1)
Tikka1_nouns = [word for word, pos in tags if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS')]
print(Tikka1_nouns)
```

# Plot number of words against how long the recipe takes to make. 
 #Do more words indicate more time to make? 
 #Can we predict how many hours a recipe will take based on the number of words? 

```python
#Number of hours each recipe takes (estimated from recipees)
    #If more time, figure out a way to find this in the text 
time=[13, 1, 5, 7, 1, 2, 1, 2, 1, 1]
type(time)
```

```python
#Count the number of words in each document 
num_words=[len(Tikka1_nouns),len(Tikka2_nouns),len(Tikka3_nouns),len(Tikka4_nouns),len(Tikka5_nouns),len(Tikka6_nouns),len(Tikka7_nouns),len(Tikka8_nouns),len(Tikka9_nouns),len(Tikka10_nouns)]
print(num_words)
type(num_words)
```

```python
#Import graphing packages 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plot number of words v time it takes to make the recipee 
plt.scatter(num_words, time)
plt.title('Number of Words vs Time to Make Tikka Masala')
plt.xlabel('Number of words')
plt.ylabel('Number of hours')
plt.show()
```

```python
#Correlation between number of words and time to make recipee? 
np.corrcoef(num_words, time)
#Correlation is 0.2343274, so very low correlation 
```

```python
#Find and plot line of best fit 
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

```python
#Reshape num_words to be compatible with numpy 
numwordstry=np.reshape(num_words, (-1, 1))
print(numwordstry)
type(numwordstry)
```

```python
#Find the line of best fit 
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(numwordstry, time) # perform linear regression
Y_pred = linear_regressor.predict(numwordstry)  # make predictions
print(Y_pred)
```

```python
plt.scatter(numwordstry, time)
plt.plot(numwordstry, Y_pred, color='orange')
plt.show()
```

```python
import math
```

```python
#R2

#SSR
ressq=sum([math.pow((x1 - x2),2) for (x1, x2) in zip(time, Y_pred)])

#SSTO
ybar= sum(time)/len(time) #3.4
ybarv=[3.4]*len(time) #array of ybar values
totsq=sum([math.pow((x1 - x2), 2) for (x1, x2) in zip(time, ybarv)])

#R2 value
Rsq=1-(ressq/totsq)
print(Rsq)
#not very good at explaining (0.0549...)
```
